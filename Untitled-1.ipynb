```python
lb=LabelEncoder()
data['Fuel_Type']=lb.fit_transform(data['Fuel_Type'])
data['Selling_type']=lb.fit_transform(data['Selling_type'])
data['Transmission']=lb.fit_transform(data['Transmission'])
data
```

```python
correlate = data.corr()
correlate
correlatewithprice = correlate['Selling_Price'].sort_values(ascending=False)
print(correlatewithprice)
sns.pairplot(correlate)
```

```python
X = data.drop(columns='Selling_Price',axis=1)
Y = data['Selling_Price']

xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.3,random_state=0)
print(xtrain.shape)
print(xtest.shape)
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)

print(model)

pred = model.predict(xtest)

print(pred)
```

```python
acc = r2_score(ytest,pred)*100
mae = mean_absolute_error(ytest,pred)
mse= mean_squared_error(ytest,pred)

print("R2 score :",acc)
print("Mean Absolute error :",mae)
print("Mean squared error :",mse)
```
